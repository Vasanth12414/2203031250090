```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample top 5 expensive cars data
top5_expensive = df_main.sort_values(by='Selling_Price', ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x='Car_Name', y='Selling_Price', data=top5_expensive, palette='viridis')
plt.title('Top 5 Most Expensive Cars by Selling Price')
plt.ylabel('Selling Price (in Lakhs)')
plt.xlabel('Car Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Price Depreciation: Selling vs Present Price
df_main['Price_Drop'] = df_main['Present_Price'] - df_main['Selling_Price']

plt.figure(figsize=(12,6))
sns.barplot(x='Car_Name', y='Price_Drop', data=df_main.sort_values('Price_Drop', ascending=False))
plt.title('Price Depreciation (Present Price - Selling Price)')
plt.ylabel('Depreciation (Lakhs)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Selling Price vs Kms Driven (Color by Fuel Type)

plt.figure(figsize=(10,6))
sns.scatterplot(x='Kms_Driven', y='Selling_Price', hue='Fuel_Type', data=df_main)
plt.title('Selling Price vs Kms Driven (by Fuel Type)')
plt.xlabel('Kms Driven')
plt.ylabel('Selling Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
